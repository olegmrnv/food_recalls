```python
# Dependencies
import json
import requests
from pprint import pprint
from config import api_key #FDA_food api needs to be inserted in config.py
import pandas as pd
from pandas.io.json import json_normalize

url = "https://api.fda.gov/food/enforcement.json?"
```

```python
# Search for articles that mention granola
query = "report_date:[20100101+TO+20191231]&limit=99"
```

```python
# Build query URL
query_url = url + "api_key=" + api_key  + "&search=" + query
print(query_url)
```

```python
# Request 99 food recalls from 2010-2019
response= requests.get(query_url).json()
pprint(response)

```

```python
# JSON to Pandas Dataframe
df = pd.DataFrame(response['results'])
df
```
